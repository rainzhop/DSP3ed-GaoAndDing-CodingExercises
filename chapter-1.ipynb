```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # 中文字体设置
```

```python
# 15
n = np.arange(-2, 30-2)
x = np.zeros(n.size)
x[2:2+6] = np.array([1, 2, 3, 4, 2, 1])
y = np.zeros(n.size)
for nn in range(30):
    y[nn] = x[nn] + 2*x[nn-2] - y[nn-1]/2.0
plt.stem(n[2:], y[2:])
plt.show()
```

```python
# 16
plt.figure(figsize=(14, 10))
n = np.arange(-2, 30-2)

# (1) 单位脉冲响应
x = np.zeros(n.size)
x[2] = 1
y = np.zeros(n.size)
for nn in range(30):
    y[nn] = 0.6*y[nn-1] - 0.08*x[nn-2] + x[nn]
plt.subplot(221)
plt.stem(n[2:], y[2:])
plt.title("(1)单位脉冲响应")

# (1) 单位阶跃响应
x = np.ones(n.size)
x[:2] = np.array([0,0])
y = np.zeros(n.size)
for nn in range(30):
    y[nn] = 0.6*y[nn-1] - 0.08*x[nn-2] + x[nn]
plt.subplot(222)
plt.stem(n[2:], y[2:])
plt.title("(1)单位阶跃响应")

# (1) 单位脉冲响应
x = np.zeros(n.size)
x[2] = 1
y = np.zeros(n.size)
for nn in range(30):
    y[nn] = 0.7*y[nn-1] - 0.1*y[nn-2] + 2*x[nn] - x[nn-2]
plt.subplot(223)
plt.stem(n[2:], y[2:])
plt.title("(2)单位脉冲响应")

# (1) 单位阶跃响应
x = np.ones(n.size)
x[:2] = np.array([0,0])
y = np.zeros(n.size)
for nn in range(30):
    y[nn] = 0.7*y[nn-1] - 0.1*y[nn-2] + 2*x[nn] - x[nn-2]
plt.subplot(224)
plt.stem(n[2:], y[2:])
plt.title("(2)单位阶跃响应")

plt.show()
```

```python
# 17
plt.figure(figsize=(14, 5))
n = np.arange(-2, 52-2)
a1, a2, b = -0.8, 0.64, 0.866

# (1) 单位脉冲响应
x = np.zeros(n.size)
x[2] = 1
y = np.zeros(n.size)
for nn in range(52):
    y[nn] = -a1*y[nn-1] - a2*y[nn-2] + b*x[nn]
plt.subplot(121)
plt.stem(n[2:], y[2:])
plt.title("单位脉冲响应h(t)")

# (1) 单位阶跃响应
x = np.ones(n.size)
x[:2] = np.array([0,0])
y = np.zeros(n.size)
for nn in range(52):
    y[nn] = -a1*y[nn-1] - a2*y[nn-2] + b*x[nn]
plt.subplot(122)
plt.stem(n[2:], y[2:])
plt.title("单位阶跃响应s(t)")

plt.show()
```

```python
# 18

n = np.arange(-2, 52-2)

x = np.zeros(n.size)
x[2] = 1

h1 = np.zeros(6)
for nn in range(6): h1[nn] = 1.0/(2**nn)
h2 = np.ones(6)
h12t = np.convolve(h1, h2, 'full')
h12 = np.zeros(n.size)
h12[2:2+h12t.size] = h12t
h3 = np.zeros(n.size)
h3[2:2+3] = np.array([1/4, 1/2, 1/4])
v = h12 + h3

y = np.zeros(n.size)
for nn in range(52):
    y[nn] = 0.9*y[nn-1] - 0.81*y[nn-2] + v[nn] + v[nn-1]

plt.figure(figsize=(21, 10))
plt.subplot(231)
plt.stem(range(6), h1)
plt.title("h1")
plt.subplot(232)
plt.stem(range(6), h2)
plt.title("h2")
plt.subplot(233)
plt.stem(n[2:], h12[2:])
plt.title("h1*h2")
plt.subplot(234)
plt.stem(n[2:], h3[2:])
plt.title("h3")
plt.subplot(235)
plt.stem(n[2:], v[2:])
plt.title("v")
plt.subplot(236)
plt.stem(n[2:], y[2:])
plt.title("整个系统的单位脉冲响应")

plt.show()
```
